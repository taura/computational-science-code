
# 練習問題: 2つの数の平均を求める

## 目標

C++ / Fortran のプログラムをコンパイルして実行する流れを確認し,
変数を使った簡単な算術式を1つ書けるようになる.

## 課題

`average.cpp` (または `average.f90`) は, 2つの数 `a` と `b` の平均を求めて表示するプログラムである.
ただし, 平均を計算する1行が `TODO` として空欄になっている.

コメント `TODO` の指示に従って **平均を計算する式を1文だけ** 書き加え,
正しい平均が表示されるようにせよ.

- C++: `double m = ...;` の右辺を埋める.
- Fortran: `m = ...` の右辺を埋める.

それ以外のコードを変更する必要はない.

ヒント: 整数同士の割り算は小数部が切り捨てられる. ここでは `a`, `b` は浮動小数点なので問題ないが,
`2` で割ることを忘れずに.

## コンパイルと実行

```
# C++
nvc++ -fast average.cpp -o average.exe

# Fortran
nvfortran -fast average.f90 -o average.exe
```

```
./average.exe
```

## 期待される結果

`a = 3.0`, `b = 4.0` の平均が表示される.

```
average of 3.000000 and 4.000000 is 3.500000
```
